### В работе исследуются

Сериальные закономерости в оптическом спектре водорода; спектр поглощения паров йода в видимой области.

### Теория
Длины волн спектральных линий водородоподобного атома описываются формулой

\begin{equation}
\dfrac{1}{\lambda_{mn}} = RZ^2 \left( \dfrac{1}{n^2} - \dfrac{1}{m^2} \right),
\end{equation}

![H](1.png)

где $R_y = 109677.6~\text{см}^{-1}$ - константа, называемая постоянной Ридберга, а $m$ и $n$ -- целые числа. 

Мы будем изучать серию Бальмера, линии которой лежат в видимой области. Для неё $n = 2$, а $m = 3,~4,~5,~6\dots$ Первые четыре линии обозначаются соответственно $H_\alpha$, $H_\beta$, $H_\gamma$, $H_\delta$.
Для молекулы йода мы рассматриваем только нулевую серию, энергетическое положение линий поглощения определяется выражением

\begin{equation}
h\nu_{0, n_2} = (E_2 - E_1) + h\nu_2 \left( n_2 + \dfrac{1}{2} \right) - \dfrac{1}{2} h \nu_1.
\end{equation}

Искать зависимость $\lambda = \lambda(\theta)$ будем в виде (дисперсионная формула Гартмана):

$\lambda = \lambda_0 + \dfrac{C}{\theta - \theta_0}.$

#### Градуировка

```python
import numpy as np
import MNK as mark
import matplotlib.pyplot as plt

predict = lambda x : 3427.07 * np.log(x) - 19698.13

# 1 - но\еоновая
# 2 - ртутная

lambdas1 = np.array([534.1, 540.1, 585.2, 594.5, 607.4, 616.4, 630.5])
degrees1 = np.array([1858,  1898,  2154,  2199, 2258, 2297, 2354])

lambdas2 = np.array([404.7, 546.1, 491.6, 579.1, 623.4])
degrees2 = np.array([851, 1934, 1515, 2123, 2337])

plt.figure(figsize=(20,16))
line1 = mark.line(lambdas1, degrees1, name = "Ртуть")
line2 = mark.line(lambdas2, degrees2, options_pt = "-r", options_ln = "r^", name = "Неон")
line1.dump("Неоновая", round_k = 1, round_b = 0)
line2.dump("Ртутная", round_k = 1, round_b = 0)

pts = np.linspace(400, 650, 1000)
y = np.array(list(map(predict, pts)))
plt.plot(pts, y, "-c", label = "Градуировка логарифмом")
plt.legend()

mark.DRAW("Градуировка", "$\lambda{}$, нм", "Показания, гр") 
```

На самом деле все точки идеально ложатся на уравнение

$$
f(x) \approx{} 314.2 \cdot{} \exp{(2.9x \cdot{} 10^{-4})}
$$

Теперь на водородной лампе найдём $H_{\alpha{}}$, $H_{\beta{}}$, $H_{\gamma{}}$

```python
graduiration = lambda x : 314.172 * np.exp(0.000290714 * x)
degrees = np.array([1103, 1474, 2521])
print("Измеренные линии водорода:", list(map(lambda x : round(x, 1), graduiration(degrees))), "+- 0.2")
degrees = np.array([2296, 2190, 1696])
print("Измеренные линии иода", list(map(lambda x : round(x, 1), graduiration(degrees))), "+- 0.2")
```

По результатам измерений водорода определим постоянную Ридберга:

* Z = 1, n = 2, m = 3, $\lambda{}_1$ = 653.8нм,
* Z = 1, n = 2, m = 4, $\lambda_2$ = 482.2нм,
* Z = 1, n = 2, m = 5, $\lambda_3$ = 432.9нм,

по формуле 

$$
R = \frac{(mn)^2}{\lambda_{mn}\cdot{}Z^2(m^2 - n^2)}
$$

Должно быть $R_y = 109677.6~\text{см}^{-1}$, что показывает более-менее неплохой способ определения постоянной Ридберга

```python
R = lambda Z, n, m, lam : ((m * n)**2) / (lam * Z**2 * (m**2 - n**2)) * 10**7
print("\nОжидаемое значение постоянной Ридберга R = 109677.6 см^{-1}\n")
print("2-3:  R =", round(R(1, 2, 3, 653.8), -2), "+- 400 см^{-1}")
print("2-4:  R =", round(R(1, 2, 4, 482.2), -2), "+- 400 см^{-1}")
print("2-5:  R =", round(R(1, 2, 5, 432.9), -2), "+- 400 см^{-1}")
```

Запишем показания спектрометра для следующих переходов в молекуле йода: $\theta_{1,0}$ - переход из первого колебательного уровня основного состояния в нулевой колебательный уровень возбуждённого состояния,  $\theta_{1,5}$ - переход из первого колебательного уровня основного состояния в пятый колебательный уровень возбуждённого состояния, $\theta_{g}$ - переход из нулевого колебательного уровня основного состояния в область непрерывного спектра возбуждённого состояния. Получаем следующие данные:

\begin{equation*}
			\theta_{1,0}=2296 \pm 1, \ \theta_{1,5}=2190 \pm 1, \ \theta_g=1696 \pm 1,
		\end{equation*}
        
откуда находим соответствующие длины волн: 

\begin{equation*}
		 	\lambda_{1,0}=(612 \pm 2) \ \text{нм}, \ \lambda_{1,5}=(594 \pm 2) \ \text{нм}, \ \lambda_g=(514 \pm 2)\ \text{нм}.
		 \end{equation*}

Определим энергию колебательного кванта возбуждённого состояния молекулы по формуле: 
		\begin{equation*}
			h \nu_2=\dfrac{h\nu_{1,5}-h \nu_{1,0}}{5}=(0.0123\pm 0.002) \ \text{эВ}
		\end{equation*}
			
Вычислим энергию электронного перехода $\Delta E=E_2-E_1$, энергию диссоциации $D_1$ в основном состоянии и энергию диссоциации $D_2$ в возбуждённом состоянии, если известно, что энергия колебательного кванта основного состояния есть $h\nu_1=0,027$эВ

$$
\Delta{}E = h\nu_{(1,0)} - \dfrac{1}{2} h\nu_2 + \dfrac{3}{2} h\nu_1 = 2.06 \pm 0.07~\text{эВ}.
$$

Тогда энергии диссоциации частиц в основном и возбуждённом состоянии, с учётом того, что энергия возбуждения атома $E_A = 0.94~\text{эВ}$:

$$
D_1 = h\nu_{\text{гр}} - E_A = 1.46 \pm 0.04~\text{эВ},
$$
$$
D_2 = h\nu_{\text{гр}} - \Delta{}E = 0.35 \pm 0.09~\text{эВ}.
$$
