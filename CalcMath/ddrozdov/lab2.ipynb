```python
import numpy as np

# возвращает коофициент матрицы в соответствии с вариантом е)
def matrix_arg_e(i, j):
  i += 1
  j += 1
  if i == j:
    return 10
  elif j == i + 1 or i > j:
    return 1 / i
  else:
    return 0

# создает строку матрицы
def get_matrix_line(index, size):
  res = [None] * size
  for j in range(size):
    res[j] = matrix_arg_e(index, j)
  return res

def create_matrix(N):
  matrix = np.array([])
  f = np.arange(stop=N+1, start=1, dtype=float)
  matrix.resize((N, N))
  for i in range(N):
    matrix[i] = get_matrix_line(i, N)
  return matrix, f


def solve_gaus_e(matrix, F):
  N = len(matrix)
  # зануляю элементы выше диагонали
  for i in range(1, N):
    k = matrix[N - i - 1][N - i] / matrix[N - i][N - i]
    matrix[N - i - 1] -= matrix[N - i] * k
    F[N - i - 1] -= F[N - i] * k
  res = [None] * N
  # по треугольной матрице вычисляю неизвестные
  res[0] = F[0] / matrix[0][0]
  for i in range(1, N):
    a = 0
    for j in range(0, i):
      a += matrix[i][j] * res[j]
    res[i] = (F[i] - a) / matrix[i][i]
  return np.array(res)


def solve_zedel_e(matrix_, f_, eps_):
  tmp_a = np.matrix(matrix_)
  n = len(tmp_a)
  V = np.zeros([n, n])
  for i in range(n):
    for j in range(i + 1, n):
      V[i, j] = tmp_a[i, j]
      tmp_a[i, j] = 0
  B = -tmp_a.I * V
  
  tmp_f = np.matrix(f_).getH()
  F = tmp_a.I * tmp_f

  # итерации
  sol = np.matrix(f_).T
  iter = 0
  while np.linalg.norm(np.array(matrix_) * np.matrix(sol) - np.matrix(f_).T) > eps_:
    iter += 1
    sol = B * sol + F
  print("num iterations: ", iter)

  res = np.array(np.zeros(n))
  for i in range(n):
    res[i] = sol[i, 0]
  return res

eps = 10e-8

matrix, f = create_matrix(100)
print("input matrix: ")
print(matrix)
print("\n")
print("f:")
print(f)
print("----------------------------------------------------------\n")

x = solve_gaus_e(np.array(matrix), np.array(f))
print("result with gauss:")
print("невязка = ", np.linalg.norm(np.array(matrix) * np.matrix(x).T - np.matrix(f).T))
print("solution:")
print(x)
print("----------------------------------------------------------\n")

iter = 2
x2 = solve_zedel_e(matrix, f, eps)
print("result with zedel:")
print("невязка = ", np.linalg.norm(np.array(matrix) * np.matrix(x2).T - np.matrix(f).T))
print("solution")
print(x2)
```
