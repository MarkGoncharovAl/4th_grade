```python
import numpy as np
import math
import matplotlib.pyplot as plt

# h[n] = 1 / 2^n, n -> [1, 21]
h = 1 / 2**np.arange(21)
x1 = 3


def diff_v1(inFunc, inH, inX):
  return (inFunc(inX + inH) - inFunc(inX)) / inH
def diff_v2(inFunc, inH, inX):
  return (inFunc(inX) - inFunc(inX - inH)) / inH
def diff_v3(inFunc, inH, inX):
  return (inFunc(inX + inH) - inFunc(inX - inH)) / (2 * inH)
def diff_v4(inFunc, inH, inX):
  return 4 * (inFunc(inX + inH) - inFunc(inX - inH)) / (3 * 2 * inH) - (inFunc(inX + 2 * inH) - inFunc(inX - 2 * inH)) / (3 * 4 * inH)
def diff_v5(inFunc, inH, inX):
  return 3 * (inFunc(inX + inH) - inFunc(inX - inH)) / (2 * 2 * inH) - 3 * (inFunc(inX + 2 * inH) - inFunc(inX - 2 * inH)) / (5 * 4 * inH) + 1 * (inFunc(inX + 3 * inH) - inFunc(inX - 3 * inH)) / (10 * 6 * inH)

def calc_func_v1(x):
  return np.sin(x**2)
def calc_func_v1_diff(x):
  return 2 * x * np.cos(x**2)

def calc_func_v2(x):
  return np.cos(np.sin(x))
def calc_func_v2_diff(x):
  return -np.sin(np.sin(x)) * np.cos(x)

def calc_func_v3(x):
  return np.exp(np.sin(np.cos(x)))
def calc_func_v3_diff(x):
  return -np.exp(np.sin(np.cos(x))) * np.cos(np.cos(x)) * np.sin(x)

def calc_func_v4(x):
  return np.log(x + 3)
def calc_func_v4_diff(x):
  return 1 / (x + 3)

def calc_func_v5(x):
  return (x + 3)**0.5
def calc_func_v5_diff(x):
  return 1 / (2 * ((x + 3)**0.5))
```

```python
plt.figure(figsize=[18, 8])
plt.grid()
plt.title("$ sin(x^2) $")
plt.ylabel('$ log(\delta, 2) $')
plt.xlabel('$ log(h, 2) $')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v1_diff(x1) - diff_v1(calc_func_v1, inH, x1)), h)), '-o', label='v1')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v1_diff(x1) - diff_v2(calc_func_v1, inH, x1)), h)), '-o', label='v2')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v1_diff(x1) - diff_v3(calc_func_v1, inH, x1)), h)), '-o', label='v3')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v1_diff(x1) - diff_v4(calc_func_v1, inH, x1)), h)), '-o', label='v4')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v1_diff(x1) - diff_v5(calc_func_v1, inH, x1)), h)), '-o', label='v5')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[18, 8])
plt.grid()
plt.title("$ cos(sin(x)) $")
plt.ylabel('$ log(\delta, 2) $')
plt.xlabel('$ log(h, 2) $')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v2_diff(x1) - diff_v1(calc_func_v2, inH, x1)), h)), '-o', label='v1')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v2_diff(x1) - diff_v2(calc_func_v2, inH, x1)), h)), '-o', label='v2')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v2_diff(x1) - diff_v3(calc_func_v2, inH, x1)), h)), '-o', label='v3')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v2_diff(x1) - diff_v4(calc_func_v2, inH, x1)), h)), '-o', label='v4')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v2_diff(x1) - diff_v5(calc_func_v2, inH, x1)), h)), '-o', label='v5')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[18, 8])
plt.grid()
plt.title("$ exp(sin(cos(x))) $")
plt.ylabel('$ log(\delta, 2) $')
plt.xlabel('$ log(h, 2) $')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v3_diff(x1) - diff_v1(calc_func_v3, inH, x1)), h)), '-o', label='v1')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v3_diff(x1) - diff_v2(calc_func_v3, inH, x1)), h)), '-o', label='v2')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v3_diff(x1) - diff_v3(calc_func_v3, inH, x1)), h)), '-o', label='v3')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v3_diff(x1) - diff_v4(calc_func_v3, inH, x1)), h)), '-o', label='v4')
plt.loglog(h, list(map(lambda inH: np.abs(calc_func_v3_diff(x1) - diff_v5(calc_func_v3, inH, x1)), h)), '-o', label='v5')
plt.legend()
plt.show()
```
